# **Análisis de datos**.



##1. Ingreso total por cada tienda.

###**Ingreso anual total por tienda.**

```python
# Lista con los DataFrames correspondientes a cada tienda.
tiendas_df = [tienda1, tienda2, tienda3, tienda4]
# Lista con los nombres de cada tienda.
tiendas_nombre = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
# Cálculo del ingreso de cada tienda.
total_tiendas = []
total_tiendas.extend(tienda.Precio.sum() for tienda in tiendas_df)
# Creación del DataFrame con los resultados.
df_tiendas = pd.DataFrame(columns=['Tiendas', 'Total'])
df_tiendas['Tiendas']= tiendas_nombre
df_tiendas['Total'] = total_tiendas
```

###**Ingreso total por mes y año**

```python
# Creación de diccionario.
ventas_anual = {}
# Iteración sobre la lista de DataFrames tiendas_df.
# Conversión de columna 'Fecha de Compra' a tipo de dato 'datetime'.
# Por cada tienda: Agrupación por 'Fecha de Compra' en formato 'año-mes'.
# Obtención de la suma de 'Precio' para obtener el total de ventas por mes.

for i, tienda in enumerate(tiendas_df, start=1):
  tiendas_df[i-1]['Fecha de Compra'] = pd.to_datetime(tiendas_df[i-1]['Fecha de Compra'], format='%d/%m/%Y')
  ventas_anual[f'Tienda {i}'] = tiendas_df[i-1][['Fecha de Compra', 'Precio']].groupby(tiendas_df[i-1]['Fecha de Compra'].dt.strftime('%Y-%m'))['Precio'].sum()

#Creación de DataFrame a partir del diccionario creado.
total_ventas_anual = pd.DataFrame(ventas_anual)
```

## 2. Ventas por categoría.

###**Número de ventas por categoría**

```python
# Obtención del número de ventas por categoría mediante la función 'value_counts'
ventas = {}

for i, tienda in enumerate(tiendas_df, start=1):
    ventas[f'Tienda {i}'] = tienda['Categoría del Producto'].value_counts()

total_categoria = pd.DataFrame(ventas)
```

###**Ingreso total por categoría de producto.**

```python
# Obtención del ingreso total, agrupando por 'Categoría del Producto' y
# Suma de 'Precio' para obtener ingreso total por 'Categoría del Producto'.
ventas_cat = {}

for i, tienda in enumerate(tiendas_df, start=1):
    ventas_cat[f'Tienda {i}'] = tienda[['Categoría del Producto', 'Precio']].groupby('Categoría del Producto').sum()['Precio']

total_ventas_cat = pd.DataFrame(ventas_cat)
```

## 3. Calificación promedio de la tienda.

```python
# Creación de lista, donde se almacena, mediante la iteración sobre la lista
# de DataFrames tiendas_df, el cálculo del promedio de la columna'Calificación',
#redondeando el resultado a dos decimales.

tiendas_mean = []
tiendas_mean.extend(round(tienda['Calificación'].mean(),2) for tienda in tiendas_df)
#DataFrame con los resultados.
df_mean = pd.DataFrame(columns=['Tiendas', 'Calificación'])
df_mean['Tiendas']= tiendas_nombre
df_mean['Calificación'] = tiendas_mean
# Ordenamiento descendente por columna 'Calificación', para mostrar
# primero a la tienda con la calificación promedio más alta.
df_mean = df_mean.sort_values('Calificación',ascending=False).reset_index(drop=True)
```

## 4. Productos más y menos vendidos.

###**Número de ventas por producto**

```python
# Obtención del número de ventas por producto mediante la función 'value_counts'
productos = {}

for i, tienda in enumerate(tiendas_df, start=1):
    productos[f'Tienda {i}'] = tienda[['Producto', 'Categoría del Producto']].value_counts()

total_productos = pd.DataFrame(productos)
total_productos = total_productos.reset_index('Categoría del Producto')
```

```python
# DataFrame con los productos con mayor y menor número de ventas por tienda,
# seleccionandolos mediante la variable global 'valor_orden'

# Primeros y últimos 5 productos.
valor_orden = 3

def prod_max_min(tienda):
  global valor_orden
  pmax = (total_productos[['Categoría del Producto', tienda]]
           .sort_values(tienda,ascending=False)
           .head(valor_orden))
  pmin = (total_productos[['Categoría del Producto', tienda]]
           .sort_values(tienda,ascending=False)
           .tail(valor_orden))
  return pd.concat([pmax,pmin])
```

```python
pmax_min_t1 = prod_max_min('Tienda 1')
pmax_min_t2 = prod_max_min('Tienda 2')
pmax_min_t3 = prod_max_min('Tienda 3')
pmax_min_t4 = prod_max_min('Tienda 4')
```

###**Ingreso total por producto**

```python
# Obtención del ingreso total, agrupando por 'Producto' y
# Suma de 'Precio' para obtener ingreso total por 'Producto'.
ventas = {}

for i, tienda in enumerate(tiendas_df, start=1):
    ventas[f'Tienda {i}'] = tienda[['Producto', 'Precio']].groupby('Producto').sum()['Precio']

total_ventas_prod = pd.DataFrame(ventas)
```

```python
# DataFrame con los productos con mayor y menor número de ingresos por tienda,
# seleccionandolos mediante la variable global 'valor_orden'

# Primeros y últimos 5 productos.
def i_max_min(tienda):
  global valor_orden
  pmax = (total_ventas_prod[[tienda]]
           .sort_values(tienda,ascending=False)
           .head(valor_orden))
  pmin = (total_ventas_prod[[tienda]]
           .sort_values(tienda,ascending=False)
           .tail(valor_orden))
  return pd.concat([pmax,pmin])
```

```python
imax_min_t1 = i_max_min('Tienda 1')
imax_min_t2 = i_max_min('Tienda 2')
imax_min_t3 = i_max_min('Tienda 3')
imax_min_t4 = i_max_min('Tienda 4')
```

## **5. Envío promedio por tienda.**

```python
# Creación de lista, donde se almacena, mediante la iteración sobre la lista
# de DataFrames tiendas_df, el cálculo del promedio de la columna
#'Costo de envío', redondeando el resultado a dos decimales.
envio_mean = []
envio_mean.extend(round(tienda['Costo de envío'].mean(), 2) for tienda in tiendas_df)
#DataFrame con los resultados.
df_envios = pd.DataFrame(columns=['Tiendas', 'Costo promedio de envio'])
df_envios['Tiendas']= tiendas_nombre
df_envios['Costo promedio de envio'] = envio_mean

df_envios_styled =df_envios.style.format({'Costo promedio de envio': '$ {:,}'})
```

#**Gráficos.**

##**Ingresos totales por tiendas**

```python
def grafico_ingresos_totales():
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax = sns.barplot(data=df_tiendas,x='Total', y=df_tiendas['Tiendas'],
                  hue=df_tiendas['Tiendas'], orient='h', palette='rocket')
  ax.set_title('Ingreso anual total por tienda', loc='left',fontsize=14, pad=20)
  ax.set_ylabel('')
  for i,j in enumerate(df_tiendas['Total']):
    ax.text(j+j*.025,i,str(f'${float(j):,}'),color='black', fontsize=9, ha='left', va='center')

  return plt.show()
```

```python
def grafico_ingresos_totales_periodo():
  fig = px.line(total_ventas_anual, x=total_ventas_anual.index,
                y=['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4'],
                title ='Ingreso total por tienda entre Enero 2020 y Marzo 2023',
                color_discrete_sequence=px.colors.qualitative.Safe, markers=True)

  fig.update_traces(line_width=2.5)

  fig.update_layout(width=1000, height = 600, xaxis={'tickangle':-45,
                                                    'tickmode': 'array',
                                                    'title': 'Fecha'},
                    yaxis_title='Total',
                    font_family='Arial', font_size=14, font_color='grey')

  fig.write_html('ingreso_total_periodo.html')
  return fig.show()
```

##**Cantidad de ventas por categoría**

```python
def grafico_ventas_categorias():
  fig = px.bar(total_categoria, x=total_categoria.index,
              y=['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4'],
              barmode='group',
              width=800,
              height=600,
              title='Número de ventas por categoría',
              labels={'variable':'Tienda', 'value':'Cantidad'},
              color_discrete_sequence=px.colors.qualitative.Safe)
  fig.update_layout(xaxis={'tickangle':-45, 'title': 'Categorías'},
                    yaxis={'title': 'Cantidad de ventas'},
                    font_family='Arial', font_size=14, font_color='grey',
                    plot_bgcolor=px.colors.qualitative.Set3[8])
  return fig.show()
```

##**Ingresos obtenidos por categoría**

```python
def grafico_ingresos_categorias():
  fig = px.bar(total_ventas_cat, x=total_ventas_cat.index,
              y=['Tienda 1','Tienda 2', 'Tienda 3', 'Tienda 4'],
              barmode='group',
              width=800,
              height=600,
              title='Ingreso total por categoría',
              labels={'variable':'Tienda', 'value':'Cantidad'},
              color_discrete_sequence=px.colors.qualitative.Safe)

  fig.update_layout(xaxis={'tickangle':-45, 'title': 'Categorías'},
                    yaxis={'title': 'Total'},
                    font_family='Arial', font_size=14, font_color='grey')
  return fig.show()
```

##**Productos con mayor y menor número de ventas e ingresos**

```python
def grafico_pmax_min(df, tienda, ax):
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  sns.barplot(data=df,x=tienda, y=df.index,
                 hue=df.index, orient='h', palette='RdYlGn_r', ax=ax)
  ax.set_title(f'Número de ventas por producto.', loc='left',fontsize=14, pad=20)
  ax.set_ylabel('')
  for i,j in enumerate(df[tienda]):
    ax.text(j,i,str(j),color='black', fontsize=9, ha='left', va='center')

  return ax
```

```python
def grafico_imax_min(df, tienda, ax):
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  sns.barplot(data=df,x=tienda, y=df.index,
                 hue=df.index, orient='h', palette='RdYlGn_r', ax=ax)
  ax.set_title(f'Ingreso total por producto', loc='left',fontsize=14, pad=20)
  ax.set_ylabel('')
  for i,j in enumerate(df[tienda]):
    ax.text(j,i,str(f'${float(j):,}'),color='black', fontsize=9, ha='left', va='center')


  return ax
```

```python
def crear_figura(df_prod, df_ing, tienda):
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
  fig.suptitle(f'{tienda}:\nMayor y menor número de ventas e ingresos por productos', fontsize = 18)

  grafico_pmax_min(df_prod, tienda, ax=ax1)
  grafico_imax_min(df_ing, tienda, ax=ax2)

  fig.tight_layout()
  plt.show()
```
